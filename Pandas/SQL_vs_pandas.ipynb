```python
import pandas as pd
import numpy as np
```

```python
retail_raw = pd.read_excel('online_retail.xlsx')
retail_raw.head()
```

```python
retail = retail_raw.copy()
retail.info()
```

```python
retail.describe()
```

```python
retail.isnull().sum()
```

```python
retail = retail.dropna()
retail.info()
```

```python
def change_float_do_str(x):
    return str(int(x))
retail.CustomerID = retail.CustomerID.apply(change_float_do_str)
retail.info()
```

```python
retail.head()
```

```python
retail = retail[retail.Quantity >=0]
```

```python
# SELECT * FROM retail WHERE CustomerID = '17850'
retail[retail['CustomerID'] == '17850']
```

```python
# SELECT * FROM retail WHERE CustomerID = '17850' and UnitPrice > 5
retail[(retail['CustomerID'] == '17850') & (retail['UnitPrice'] > 5)]
```

```python
# SELECT * FROM retail WHERE CustomerID = '17850' or Country = 'France';
retail[(retail['CustomerID'] == '17850') | (retail['Country'] == 'France')]
```

```python
# SELECT * FROM retail WHERE InvoiceNo is not null;
retail[retail['InvoiceNo'].notnull()]
```

```python
# SELECT CustomerID, count(*) FROM retail GROUP BY CustomerID; -> Ile razy konkretne ID Klienta pojawiło
# Się w naszej ramce danych

retail.groupby('CustomerID').size()
```
