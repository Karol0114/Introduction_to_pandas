```python
import pandas as pd
```

```python
df_raw = pd.read_csv('appstore_games.csv')
df_raw.head()
```

```python
# Wstepne przygotowywanie danych:

df_raw.columns
```

```python
# Tworzymy kopię

df = df_raw.copy()
df.info()
```

```python
# Usuwamy niepotrzebne kolumny:
df = df.drop(columns = ['URL', 'Icon URL', 'Subtitle', 'Description', 'In-app Purchases'])
df.head()
```

```python
# ustalamy index 
df = df.set_index('ID')
df
```

```python
# Sprawdźmy statystyki

df.describe().T
```

```python
df.describe(include = ['object']).T
```

```python
df.columns
```

```python
# Stanaryzujemy nazyw kolumn spacje zasętpujemy '_', i nazwy z małych liter

df.columns = [col.lower() for col in df.columns]
df.head()
```

```python
df.columns = [col.replace(' ', '_') for col in df.columns]
df.head()
```
